## Condition Probabilities

Its just simply the probability of something happend knowing something.

For example

What is the probability of someone do not love candy and love soda, knowing that that person already loves soda.

   **p(not love c and loves s | loves s)** 
   
    - population of people that love soda is 7 and not the whole population. 
    - population taht do not love candy inside population loves soda are 5,so in that case are 5/7 = 0.71

### Bayes Formula

A = does not love candy
B = loves soda

P(A & B | B) = ( P(A & B | A) P(A) ) / P(B)

P(A & B | A) = ( P(A & B | B) P(B) ) / P(A)
